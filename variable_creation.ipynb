# Set Basic Parameters for the Following Process

unique_id is the vector that contains all unique order IDs. Each row corresponds to a specific order ID

```python
unique_id = item['order_id'].unique()
```

count is the vector having the same size as unique_id and currently have all 0

column is the vector having the same length as item and currently have all 0

# Action Count [t,s]

We need to add 20 more columns to identify the action count in each time interval

```python
interval = item[['order_id', 'fraction']]
interval
```

```python
def calculate_interval_count(min, max):
    for i in range(unique_id.shape[0]):
        tmp = interval[interval['order_id'] == unique_id[i]]
        index = (min <= tmp['fraction']) & (tmp['fraction'] < max) 
        count[i] = np.sum(index)
        
    for i in range(unique_id.shape[0]):
        index = arr['order_id'] == unique_id[i]
        column[index] = count[i]
    return column
```

```python
min = 0
max = 0.05
name = ['0-0.05', '0.05-0.1', '0.1-0.15', '0.15-0.2', '0.2-0.25', '0.25-0.3', '0.3-0.35', '0.35-0.4', '0.4-0.45', '0.45-0.5',
        '0.5-0.55', '0.55-0.6', '0.6-0.65', '0.65-0.7', '0.7-0.75', '0.75-0.8', '0.8-0.85', '0.85-0.9', '0.9-0.95', '0.95-1']
for i in range(20):
    column = calculate_interval_count(min, max)
    min = max
    max = max + 0.05
    interval[name[i]] = column
interval
```

# Consign Count, Receive Count, Arrive Count, Depart Count, Scan Count

```python
def fn(action_name):
    arr10 = item[['order_id', 'action']]
    for i in range(unique_id.shape[0]):
        tmp10 = arr10.loc[arr10['order_id'] == unique_id[i]]
        count[i] = np.sum(tmp10['action'] == action_name)
    column10 = np.zeros(arr10.shape[0])
    for i in range(unique_id.shape[0]):
        index10 = arr10['order_id'] == unique_id[i]
        column10[index10] = count[i]
    return column10
```

```python
column = fn('SIGNED')
item['recieve_count'] = column
```

```python
column = fn('ARRIVAL')
item['arrive_count'] = column
```

```python
column = fn('DEPARTURE')
item['depart_count'] = column
```

```python
column = fn('SENT_SCAN')
item['scan_count'] = column
```

```python
column = fn('CONSIGN')
item['consign_count'] = column
```
